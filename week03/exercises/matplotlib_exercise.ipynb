```python
# compuet the x and y coordinates for points on a sine curve
x = np.arange(0, 4 * np.pi, 0.1) # 0.1 is the stepping point
y = np.sin(x)
plt.title("Sine Wave Form")

plt.plot(x, y)
plt.show()
```

```python
# subplot()
#Compute  the x and y coordinates or pointson sine and cosine function

x = np.arange(0, 5*np.pi, 0.1)
y_sine = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
plt. subplot(2, 1, 1)

# Make the first plot
plt.plot(x, y_sine)
plt.title("Sine")
plt.show()

# Set the second subplot as active, and make the second plot.
plt.subplot(2,1,2)
plt.plot(x, y_cos, 'g--')
plt.title("Cosine")
plt.show()
```

```python
## Bar plot

x = [2, 8, 10]
y = [11, 6, 9]

x2=[3, 9, 11]
y2= [6, 15, 7]
plt.bar(x, y)
plt.bar(x2, y2, color='g')
plt.title("Bar graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

# Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

#rectangular box plot
plt.boxplot(data, vert=True, patch_artist=True)
plt.show()
```

# Pie Chart

```python
# Data to plot

labels= ['python', 'C++', 'Ruby', 'Javascript']
sizes = [215, 130, 245, 210]
colors= ['Red', "yellowgreen", "Orange", "blue"]
explode = (0.1, 0, 0, 0)

#plot
plt.pie(sizes, explode=explode, labels = labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis("equal")
plt.show()
```
